```python
import pandas as pd
import os
```

```python
# Pfad zum Ordner "data"
data_path = "data"

# Liste aller CSV-Dateien im Ordner
files = [f for f in os.listdir(data_path) if f.endswith(".csv")]

# Alle CSVs einlesen
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(data_path, file))
    dfs.append(df)

# Optional: anzeigen, welche Dateien gefunden wurden
files
```

```python
for file, df in zip(files, dfs):
    print("=== Datei:", file, "===")
    print(df.columns.tolist())
    print()
```

```python
df_all = pd.concat(dfs, ignore_index=True)
df_all.tail()
```

```python
df_all.shape
```

```python
df_all["MessungDatZeit"] = pd.to_datetime(df_all["MessungDatZeit"])
sorted(df_all["MessungDatZeit"].dt.year.unique())
df_all["MessungDatZeit"].dt.year.value_counts().sort_index()
```

```python
df_all = df_all.drop(columns=["D1ID", "D2ID", "D3ID", "D4ID", "LieferDat"])
df_all.head()
```

```python
df_all = df_all.rename(columns={
    "MSID": "measurement_site_id",
    "MSName": "measurement_site_name",
    "ZSID": "counting_site_id",
    "ZSName": "counting_site_name",
    "Achse": "axis",
    "HNr": "house_number",
    "Hoehe": "position_description",
    "EKoord": "east_coordinate",
    "NKoord": "north_coordinate",
    "Richtung": "direction",
    "Knummer": "signal_id",
    "Kname": "signal_name",
    "AnzDetektoren": "num_detectors",
    "MessungDatZeit": "timestamp",
    "AnzFahrzeuge": "vehicle_count",
    "AnzFahrzeugeStatus": "vehicle_count_status"
})
```

```python
# Replace values in measurement_site_name and position_description
df_all["measurement_site_name"] = df_all["measurement_site_name"].replace({
    "Unbekannt": "Unknown"
})

df_all["position_description"] = df_all["position_description"].replace({
    "Unbekannt": "Unknown"
})

# Replace direction values
df_all["direction"] = df_all["direction"].replace({
    "einwärts": "inbound",
    "auswärts": "outbound",
    "einwärts (ohne Busse)": "inbound (excluding buses)"
})

# Replace vehicle_count_status values
df_all["vehicle_count_status"] = df_all["vehicle_count_status"].replace({
    "Gemessen": "Measured",
    "Fehlend": "Missing",
    "Imputiert": "Imputed"
})
```
